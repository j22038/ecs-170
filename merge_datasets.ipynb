## Get Train Test Split

```python
import os

input_directory = 'data/raw/PlantDiseasesDataset'
class_sizes = {} # class_sizes[class_name] = [training size, test size, class name]
for class_name in os.listdir(input_directory):
    class_path = os.path.join(input_directory, class_name)
    
    train_path = os.path.join(class_path, 'train')
    test_path = os.path.join(class_path, 'valid')

    for sub_class_name in os.listdir(train_path):

        # count training images
        train_count = 0
        sub_class_train_path = os.path.join(train_path, sub_class_name)

        for current_address, directory, file_name in os.walk(sub_class_train_path):
            for individual_files in file_name:
                train_count += 1

        class_sizes[sub_class_name] = [train_count]

    for sub_class_name in os.listdir(test_path):

        # count test images
        test_count = 0
        sub_class_test_path = os.path.join(test_path, sub_class_name)

        for current_address, directory, file_name in os.walk(sub_class_test_path):
            for individual_files in file_name:
                test_count += 1

        class_sizes[sub_class_name].append(test_count)
        class_sizes[sub_class_name].append(class_name)

# print sizes
for class_name in class_sizes:
    print(f'{class_name}, training size: {class_sizes[class_name][0]}, test size: {class_sizes[class_name][1]}')
print()

# print ratios
for class_name in class_sizes:
    total = class_sizes[class_name][0] + class_sizes[class_name][1]
    train_ratio = class_sizes[class_name][0] / total
    test_ratio = class_sizes[class_name][1] / total
    print(f'{class_name}: \n training: {train_ratio} \n testing: {test_ratio}')
```

## Copy Files To Destination

```python
import shutil

source_directory = 'data/raw/PlantVillage'
dest_directory = 'data/raw/PlantDiseasesDataset'
for class_name in os.listdir(source_directory):

    class_path = os.path.join(source_directory, class_name)
    for current_address, directory, file_name in os.walk(class_path):
        total = class_sizes[class_name][0] + class_sizes[class_name][1]
        train_ratio = class_sizes[class_name][0] / total
        test_ratio = class_sizes[class_name][1] / total

        training_size = int(train_ratio * len(file_name))
        training_files = file_name[:training_size]
        training_dest = os.path.join(dest_directory, class_sizes[class_name][2], 'train', class_name)
        for file_name in training_files:
            training_source = os.path.join(class_path, file_name)
            shutil.copy2(training_source, training_dest)

        # test_size = int(test_ratio * len(file_name))
        test_files = file_name[training_size:]
        test_dest = os.path.join(dest_directory, class_sizes[class_name][2], 'valid', class_name)
        for file_name in test_files:
            test_source = os.path.join(class_path, file_name)
            shutil.copy2(test_source, test_dest)        
```
